You have the task to find the average age of the residents living in your state. You can choose find out the ages of a few people living around your neighbourhood. Once you collect that data, you may be tempted to calculate the mean of the data you collected $\bar{x}$ (**sample mean**) and call it the mean of the entire population of your state $\mu$(**population mean**). Let us ignore the fact, that you have too small a sample for correctly guessing anywhere near the true population mean. The sample mean in this case is called a **point estimate**. This estimate may or may not be the true **population mean**. 

To increase your reliability in your estimate, you could make use of a **confidence interval**. A confidence interval is a statement of the form $l \leq \mu \leq u$. It is important to note that this statement in about the population parameter. The characteristics of the drawn sample is used to find the $l$,$u$, i.e. the **upper and lower confidence intervals**

Before doing the simulation, you may ask: *What is an appropriate proabability distribution to model the age spread of the population?* Even though age is a continuous variable, we can use discrete values (with binning) to analyse it. Let us use the poisson RV (for no particular reason)

```python
population = stats.poisson.rvs(mu=10,size=500000,loc=30)
```

```python
fig,ax = plt.subplots()
ax.hist(population)
```
