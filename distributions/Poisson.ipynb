```python
import numpy as np
import pandas as pd
import seaborn as sns
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
population = stats.poisson.rvs(mu=10,size=500)
```

```python
fig,ax = plt.subplots(figsize=(6,4))
ax.hist(population)
plt.show()
```

```python
print(population.mean())
print(population.var())
```

```python
population = stats.poisson.rvs(loc=10,mu=10,size=500)
fig,ax = plt.subplots(figsize=(6,4))
ax.hist(population)
plt.show()
```

```python
print(population.mean())
print(population.var())
```
