```python
import numpy as np
import matplotlib.pyplot as plt
```

# bernoulli RVs

```python
n = 1 # for bernoulli
p = 0.3
var = p*(1-p)
size_arr = np.arange(1,10000,99)
```

## lets see how quickly the sample mean converges to population mean

```python
res = {}
for size in size_arr:
    p_hat = np.mean(np.random.binomial(n,p,size))
    res[size] = p_hat
```

```python
sizes = list(res.keys())
p_hats = list(res.values())
standard_error = np.sqrt(var/np.array(sizes))
```

```python
fig,axs = plt.subplots(1,2,figsize=(12,4))
axs[0].plot(sizes[1:],p_hats[1:],label="p_hat")
axs[0].plot(sizes[1:],p*np.ones((len(sizes[1:]))),label="p")
axs[0].legend()

axs[1].plot(sizes[1:],standard_error[1:],label="standard_error")
axs[1].legend()

axs[0].set(xlabel="n")
axs[1].set(xlabel="n")
```

As the sample size increases, the sample mean comes closer to population mean

## how does sample variance vary with the size?

```python
res_v = {}
for size in size_arr:
    var_hat = np.var(np.random.binomial(n,p,size))
    res[size] = var_hat
```

```python
sizes = list(res.keys())
var_hats = list(res.values())
```

```python
fig,ax = plt.subplots(figsize=(6,4))
ax.plot(sizes[1:],var_hats[1:],label="var_hat")
ax.plot(sizes[1:],var*np.ones(len(sizes[1:])),label="var")
ax.legend()
```

Why is the sample variance always lesser than population variance for **p=0.5**. This is a peculiar property of the variance of a binomial variable. It is maximized at value p = 0.5. Since sample p_hat is not equal to p in most cases, p_hat(1-p_hat) is lesser than 0.25 always

# Binomial Random Variables

```python
# n_list =  np.arange(1,1000,10)
n = 10
p = 0.3
mu = n*p
var = n*p*(1-p)
sizes = np.arange(1,1001,10)
```

```python
np.mean(np.random.binomial(10,0.5,10))
```

```python
res = {"size":[],"p_hat":[],"mu_hat":[],"var_hat":[],"mu":[],"var":[]}
for size in sizes:
    
    n_success = np.mean(np.random.binomial(n,p,size))
    p_hat = n_success/n
    mu_hat = n_success
    var_hat = n*p_hat*(1-p_hat)

   
    res["size"].append(size)
    res["p_hat"].append(p_hat)
    res["mu_hat"].append(mu_hat)
    res["mu"].append(mu)
    res["var_hat"].append(var_hat)
    res["var"].append(var)
```

```python
size = np.array(res["size"])
mu = np.array(res["mu"])
mu_hat = np.array(res["mu_hat"])
var = np.array(res["var"])
var_hat = np.array(res["var_hat"])
```

```python
fig,axs = plt.subplots(1,2,figsize=(12,4))
axs[0].plot(size[1:],mu[1:],label="mu")
axs[0].plot(size[1:],mu_hat[1:],label="mu_hat")
axs[0].legend()

axs[1].plot(size[1:],var[1:],label="var")
axs[1].plot(size[1:],var_hat[1:],label="var_hat")
```
